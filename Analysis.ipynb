### Finding major topics of movies in the dataset using LDA.

```python
import gensim
from gensim.utils import simple_preprocess
from gensim.parsing.preprocessing import STOPWORDS
```

```python
def preprocess(text):
    result=[]
    for token in gensim.utils.simple_preprocess(text) :
        if token not in gensim.parsing.preprocessing.STOPWORDS and len(token) > 3:  ## removes stop-words
            result.append(token)       
    return result
```

Conjucture here is that all the topics name will be somehow included in the column keywords.

```python
gf.keywords = gf.keywords.apply(lambda x: x.replace('-', ' '))
```

```python
X = gf['keywords']
```

```python
processed_docs = []

for doc in X: 
    processed_docs.append(preprocess(doc))
```

```python
processed_docs[0]
```

```python
dictionary = gensim.corpora.Dictionary(processed_docs) # makes the dictionray of the words that is it converts it assign an index to each word
```

```python
dictionary.filter_extremes(no_below=15, no_above=0.1, keep_n= 30000)
bow_corpus = [dictionary.doc2bow(doc) for doc in processed_docs]
```

```python
lda_model =  gensim.models.LdaMulticore(bow_corpus,num_topics = 14, id2word = dictionary, passes = 10, workers = 3)
```

```python
for idx, topic in lda_model.print_topics(-1):
    print("Topic: {} \nWords: {}".format(idx, topic ))
    print("\n")
```
